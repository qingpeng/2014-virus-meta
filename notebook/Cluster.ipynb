```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def scatter_3d(coord_matrix, n,title="", axis1=0, axis2=1, axis3=2):
    # adapted from matplotlib gallery here:
    # http://matplotlib.org/examples/mplot3d/scatter3d_demo.html
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    xs = coord_matrix[axis1]
    
    ys = coord_matrix[axis2]
    zs = coord_matrix[axis3]
    print xs,ys,zs
    plot = ax.scatter(xs, ys, zs,)

    ax.set_xlabel('PC %d' % (axis1 + 1))
    ax.set_ylabel('PC %d' % (axis2 + 1))
    ax.set_zlabel('PC %d' % (axis3 + 1))
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    ax.set_zticklabels([])
#    fig.colorbar(plot)
    ax.set_title(title)
    
    for i, txt in enumerate(n):
        print i,txt
        ax.annotate(txt, (xs[i],ys[i],zs[i]))
    
    plt.show()
```
