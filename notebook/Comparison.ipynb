### similarity matrix and clustering

Coverage of a read in different samples

If the median k-mer frequency of a read in a sample is >=1, we consider this read as being covered by that sample.

More stringent in Compareads(Maillet et al. 2012): Given integers k and t, two sequences s1 and s2 are said similar if and only if they share at least t non-overlapping k-mers.

```python
!more AB.tr.fq.trimmed.fq.comb.head.txt
```

### Distance

- AinB: number of reads in sample A that are covered in sample B
- BinA: number of reads in sample B that are covered in sample A
- A: number of reads in sample A
- B: number of reads in sample B

distance = (AinB+BinA)/(A+B)

### digital normalization --> 5x coverage.  

- The purpose is to decrease the influence of uneven species abundance distribution. 
- Previous comparison is the comparison of genomic contents -- including everything, all the DNAs in samples that are sequenced,  with high abundance or low abundance.
- This comparison is close to the comparison on species level without the influence of species abundance.
- Now the abundance/coverage is in the range of 1 to 5.
